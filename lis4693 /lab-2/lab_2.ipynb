```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
import requests
url = "https://raw.githubusercontent.com/scarlettpob/lis4693/04cc285e861191591807e577505cc43c03f0bc91/lab-1/transcript.txt"
response = requests.get(url)
text = response.text
print(text)
```

```python
#TASK 2
print("Number of characters:", len(text)) # print number of characters

print(text[:100])   # print first 100 characters
```

```python
#TASK 3
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")

doc = nlp(text)

sentences = list(doc.sents)

print("Number of sentences:", len(sentences))

print("\nFirst 5 sentences:")
for sent in sentences[:5]:
    print(sent)
words = [token.text.lower() for token in doc if token.is_alpha]

```

```python
#TASK 4
print("Total words:", len(words))
print("Unique words:", len(set(words)))
```

```python
#TASK 5
from collections import Counter

word_freq = Counter(words)

print("Top 10 most frequent words:")
for word, count in word_freq.most_common(10):
    print(word, count)
```

```python
#TASK 6

nlp2 = spacy.load("en_core_web_sm")

doc2 = nlp2(text)

print("Named Entities:")
for ent in doc2.ents:
    print(ent.text, ent.label_)
```

```python
#TASK 7
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp2.vocab, attr="LOWER")

phrases = ["Jo March", "Laurie Laurence", "marriage proposal"]

patterns = [nlp.make_doc(phrase) for phrase in phrases]
matcher.add("MOVIE_PHRASES", patterns)
matches = matcher(doc2)

print("Matches found:")
for match_id, start, end in matches:
    print(doc2[start:end])
    print("Sentence:", doc2[start].sent)
```

Loading the transcript directly from GitHub worked smoothly, and spaCy’s tools made sentence segmentation and entity recognition straightforward. The PhraseMatcher was especially helpful in identifying topic-related phrases.
