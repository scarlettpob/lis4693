```python
import pandas as pd
import altair as alt
from vega_datasets import data as vega_data
import requests
import io

url = "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"
response = requests.get(url)
response.raise_for_status() # Raise an exception for HTTP errors
text = response.text
seattle_df = pd.read_csv(io.StringIO(text))
seattle_df.head()
```

```python
import altair as alt

# Visualization 1: Line chart for total checkouts per year
# Aggregate data to get total checkouts for each year
annual_checkouts = seattle_df.groupby('CheckoutYear')['Checkouts'].sum().reset_index()

chart_line = alt.Chart(annual_checkouts).mark_line(point=True).encode(
    x=alt.X('CheckoutYear:O', title='Checkout Year'),
    y=alt.Y('Checkouts', title='Total Number of Checkouts'), # Added y-axis label for clarity
    tooltip=['CheckoutYear', 'Checkouts']
).properties(
    title='Total Checkouts Over Years'
).interactive() # Added .interactive() for panning and zooming

chart_line
```

```python
# Visualization 2: Circle chart for total checkouts per month
# Aggregate data to get total checkouts for each month
monthly_checkouts = seattle_df.groupby('CheckoutMonth')['Checkouts'].sum().reset_index()

chart_circle = alt.Chart(monthly_checkouts).mark_circle(size=100, color='steelblue').encode(
    x=alt.X('CheckoutMonth:O', title='Checkout Month'),
    y=alt.Y('Checkouts', title='Total Number of Checkouts'),
    tooltip=['CheckoutMonth', 'Checkouts']
).properties(
    title='Total Checkouts by Month'
).interactive() # Added .interactive() for panning and zooming

chart_circle
```
